# Exploratory Data Analysis (EDA)

```python
df.columns
```

**Gender**

```python
df["Gender"].unique() # it will convert into numpy array
```

```python
# Gender Count (Male, Female)
sns.countplot(x = "Gender", data = df, hue = "Gender")
plt.show()
# palette="cool"
```

```python
ax =sns.countplot(x = "Gender", data = df, hue = "Gender")
plt.title("Bar Graph: Male vs. Female Customer Counts")
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("Gender 1", dpi = 100)
plt.show()
```

**number of Female and Male customers are 7832 and 3407 respectively.** 

```python
df.groupby(["Gender"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
```

```python
sales_gen = df.groupby(["Gender"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
sns.barplot(x = "Gender", y = "Amount", data = sales_gen, hue = "Gender")
plt.title("Gender and Sales Amounts")
plt.savefig("Gender Sales.png")
plt.show()
```

**From above graphs we can see that most of the buyers are Females and even the purchasing power of Female is greater than Male**

**State**

```python
# sates included in dataset
print(df["State"].unique())
len(df["State"].unique())
```

** This Dataset has sales from 16 States. **

```python
plt.rcParams["figure.figsize"] = [22,7]
# changing shape of the figure, default figure size of plot is (6.4, 4.8)

# number of sales Statewise
ax = sns.countplot(x = "State", data = df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("State and number of sales.png")
plt.show()
```

```python
# Total amount and states
df.groupby(["State"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
```

```python
sns.set(rc = {'figure.figsize' : (25,7)})

sales_state = df.groupby(["State"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
sns.barplot(x = "State", y = "Amount", data = sales_state, hue = "State")
plt.savefig("States and sales amount.png")
plt.show()
```

```python
# Total orders and states
df.groupby(["State"], as_index = False)["Orders"].sum().sort_values(by = "Orders", ascending = False)
```

```python
# total number of orders from top 10 states
orders_state = df.groupby(["State"], as_index = False)["Orders"].sum().sort_values(by = "Orders", ascending = False)
sns.barplot(x = "State", y = "Orders", data = orders_state, hue = "State")
plt.savefig("top 10 states.png")
plt.show()
```

**from above graphs we can see that unexpectedly most of the orders are from Uttar Pradesh, Maharashta and karnataka respectively and total
sales/amount is from Uttar Pradesh, Maharashta and karnataka.**
